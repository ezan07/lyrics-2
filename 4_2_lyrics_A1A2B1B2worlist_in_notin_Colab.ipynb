```python

#_lyrics

class lyricstextedit:
    
    def __init__(self,lyrics):
        
        #_read
        
        f1 = open(lyrics)
        r1 = f1.read()
        
        #_print("12_read=",r1)
        print("14_Number_of_characters_len(read)=",len(r1))
        #_read_len=1
        f1.close()
        
        #_replace
        
        r2= r1.replace(" ", '\n')
        #_print("20_replace=",r2)
        print("22_Number_of_characters_len(replace)=",len(r2))

        #_split
        
        r3 = r2.split("\n")
        #_print("28_split=",r3)
        print("28_len(split_lyrics)=",len(r3))
        
        #_Element of the set , duplicate deletion
        #_Top of list_' '_First character
        #_set
        
        r4=list(set(r3))
        #_print("36_list(set(r3))=",r4)
        print("36_duplicate_deletion_len(set(split(lyrics)))=",len(r4))
        
        #_sorted
        
        self.sortedr4 = sorted(r4)
        print("41_self.sortedr4[0],[1]=",self.sortedr4[0],",",self.sortedr4[1])
        print("42_self.sortedr4[2],[3]=",self.sortedr4[2],",",self.sortedr4[3])
        print("43_len(sorted(set(word)))=",len(self.sortedr4))
        
        
        #_join → one_word
        
        halfsizejoinr4 = ' '.join(self.sortedr4)
        print("49_len(halfsize_join_word)Number_of_characters =",len(halfsizejoinr4))
          
        #_write (one_word)
        
        lyrics_r4word = lyrics.replace(".txt","_harfsizespacejoin.txt")
        print("54_write_text=",lyrics_r4word)
        
        f1 = open(lyrics_r4word,"w")
        f1.write(halfsizejoinr4)
        
        f1.close
        
        #_read (one_word)
        
        f1 = open(lyrics_r4word,"r")
        f2 = f1.read()
        #_print("73_halfsizejoinr4.txt=",f2)
        print("66_read_len(lyrics_r4.txt)Number_of_characters=",len(f2),"\n")
        f1.close
        
        
#_in_list,not_in_list
#_dictionary：A1,A1A2,A1A2B1,A1A2B1B2

class wordinclude:
    
    def __init__(self,ABname,dictitle,lyrics):
        
        #_read
        
        AB_name = open(ABname,"r")   #_len=1085
        ABdic = AB_name.read()
        #_print("95_AB_name=",ABdic)
        print("82_ABdic_len(Number_of_characters)=",len(ABdic))
        
        #_list = [abc,---,---,--]_element_comparison
        #_split
        
        ABdicsplit = ABdic.split(" ")        
        #_print("102_split(dictionary)=",ABdicsplit)    #_= ['', "'m", "'re", "'s", 'A.M.',
        print("89_len(dictionary_word)=",len(ABdicsplit)) 
        
        AB_name.close
        
        #_Instance variable name
        #_lyrics_ˈlɪrɪks_
        
        lyrics_word = lyedit.sortedr4
        #_print("111_lyrics_word=",lyrics_word)
        print("98_len(lyrics_word)=",len(lyrics_word))
        
        lywl_range = list(range(len(lyrics_word)))
        print("101_len(range(len(lyrics_word)))=",len(lywl_range),"\n")
        
        #_c.append
        
        c1=[]    #_in_list
        c2=[]    #_not_in_list
        
        lyrics_notin = lyrics.replace(".txt","_notin.txt")
        lyrics_in = lyrics.replace(".txt","_in.txt")
        
        f1 = open(lyrics_notin,"w")
        f2 = open(lyrics_in,"w")
        f1.close
        f2.close
        
        for i in lywl_range:
            
            if lyrics_word[i] not in ABdicsplit:
                
                #_print("131_notin_lyrics_word[",i,"] =",lyrics_word[i])
                c1.append(lyrics_word[i])
                
            else:             #_listed_in
                #_print("135_in_lyrics_word[",i,"] =",lyrics_word[i])
                c2.append(lyrics_word[i])
                
        #_print("138_notin_c1=",c1,"\n")
        #_     c1= ['heaven', 'hell', 'countries',,,]
        
        print("130_",dictitle,"_notin_word_len(c1)=",len(c1))
        
        #_print("143_in_c2=",c2,"\n")
        #_    c2= ['imagine', 'there', 'is',,,,]
        
        print("135_",dictitle,"_in_word_len(c2)=",len(c2))       
        
        #_join()
        
        self.c1 = c1
        c11= " ".join(self.c1)
        print("141_",dictitle,"_not_word_len(join_c1)Number_of_characters_self.c11_write=",len(c11))
        
        #_write
        
        f1.write(c11)
        
        #_join()
        self.c2 = c2
        c12= " ".join(self.c2)
        print("150_",dictitle,"_in_word_len(join_c2)Number_of_characters_self.c12_write=",len(c12),"\n")
        
        #_write
        
        f2.write(c12)
        
        
#_distribution

class wordABdistribution:
    
    def __init__(self):
              
        #_class_call（lyrics_name）
        
        lyricsword = lyedit.sortedr4       
        print("166_lyricsword[0]",lyricsword[0],",_[1]",lyricsword[1],",_[2]",lyricsword[2],",_[3]",lyricsword[3],",_[4]",lyricsword[4])
        #_['', 'a', 'allowed', 
        print("168_len(lyricsword)=",len(lyricsword),"\n")
        
        #_comparison
        #_self.c1_(not_in_list) / self.c2_(in_list)
        
        #_A1
        
        dicname = "/content/drive/My Drive/Colab/A1_1200UTF8_wordlist.txt"
        dictitle = "A1"
        
        #_class(-)instance
        
        w_inc = wordinclude(dicname,dictitle,lyrics)
        
        #_in_list
        
        A1in = w_inc.c2
        print("185_A1in_len=",len(A1in),"\n")
        #_print("201_A1in",A1in)
        
        #_A1A2
        
        dicname = "/content/drive/My Drive/Colab/A1A2_2370UTF8_wordlist.txt"
        dictitle = "A1A2"
        w_inc = wordinclude(dicname,dictitle,lyrics)
        
        #_in_list
        
        A1A2in = w_inc.c2
        print("197_A1A2in_len=",len(A1A2in),"\n")
        #_print("211_A1A2in",A1A2in)
        
        #_A1A2B1
        
        dicname = "/content/drive/My Drive/Colab/A1A2B1_4549UTF8_wordlist.txt"
        dictitle = "A1A2B1"
        w_inc = wordinclude(dicname,dictitle,lyrics)
        
        #_in_list
        
        A1A2B1in = w_inc.c2
        print("209_A1A2B1in_len=",len(A1A2B1in),"\n")
        #_print("220_A1A2B1in",A1A2B1in)
        
        #_A1A2B1B2
        
        dicname = "/content/drive/My Drive/Colab/A1A2B1B2_7046UTF8_wordlist.txt"
        dictitle = "A1A2B1B2"
        w_inc = wordinclude(dicname,dictitle,lyrics)
        
        #_not_in_list
        
        A1A2B1B2notin = w_inc.c1
        print("221_A1A2B1B2notin_len=",len(A1A2B1B2notin))
        #_print("230_A1A2B1B2notin",A1A2B1B2notin)
        
        #_in_list
        
        A1A2B1B2in = w_inc.c2
        print("227_A1A2B1B2in_len=",len(A1A2B1B2in))
        #_print("234_A1A2B1B2in=",A1A2B1B2in)
        #_print("235_set(A1A2B1B2in)=",set(A1A2B1B2in))
        
        #_A2add_(set_difference)
        
        A2add = set(A1A2in) - set(A1in)
        print("234_A2add=",A2add)
        print("235_len(A2add)=",len(A2add))
        
        #_B1add_(set_difference)
        
        B1add = set(A1A2B1in) - set(A1A2in)
        print("240_B1add=",B1add)
        print("241_len(B1add)=",len(B1add))
        
        #_B2add_(set_difference)
        
        B2add = set(A1A2B1B2in) - set(A1A2B1in)
        print("246_B2add=",B2add)
        print("247_len(B2add)=",len(B2add))
        
        #_lyrics_wordAB_(sum)
        
        lyrics_wordAB = set(A1in)|A2add|B1add|B2add|set(A1A2B1B2notin)
        lwAB = list(lyrics_wordAB)
        print("253_AB(in_add_notin)union=\n",lwAB)
        print("254_len(AB(in_add_notin)union)=",len(lwAB),"\n")
            
        c3 = []
        c4 = []
        a = list(range(len(lyricsword)))
        
        for i in a:
            if lyricsword[i] in lwAB:
                c3.append(lyricsword[i])
            else:
                c4.append(lyricsword[i])
                
        print("266_lyricsword_in_dic=\n",c3)
        print("267_len(lyricsword_in_dic)=",len(c3),"\n")
        
        print("269_lyricsword_not_in_dic=",c4)
        print("270_len(lyricsword_not_in_dic)=",len(c4),"\n")
        
        
#_lyrics

lyrics = "/content/drive/My Drive/Colab/imagine.txt"

#_lyrics = "blowin in the wind.txt"

#_class

lyedit = lyricstextedit(lyrics)

#_class

wABdis = wordABdistribution()


print()
#
```
